```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret,image=cap.read()
    cv2.imshow("hello",image)
    if cv2.waitKey(100) == 13:
        break
cv2.destroyAllWindows()
```

```python
def gk(e,x,y,w,h):
    if cv2.EVENT_MOUSEMOVE == e:
        cv2.circle(image,(x,y),50,[255,255,0],-1)
        print('x:',x,' y:',y)


while True:
    ret,image=cap.read()    
    cv2.setMouseCallback("hii", gk)
    cv2.imshow("hii",image)

    if cv2.waitKey(100) == 13:
        break
cv2.destroyAllWindows()
```
