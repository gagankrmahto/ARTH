```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
ret,img = cap.read()
```

```python
pic_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
pic_rgb = cv2.resize(pic_rgb,(512,512))
plt.imshow(pic_rgb)
```

```python
cap.release()
```

```python
pic2 = cv2.imread('img.png')
pic2 = cv2.resize(pic2,(512,512))
pic2 = cv2.cvtColor(pic2 ,  cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(pic2)
```

```python
blended = cv2.addWeighted(src1=pic_rgb , alpha=0.8,  src2=pic2, beta=0.2, gamma=0)
```

```python
plt.imshow(blended)
```
